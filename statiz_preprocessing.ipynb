```python
import pandas as pd

# 예제 데이터프레임 생성
data = {'A': [1, 2, 3, 1],
        'B': [5, 6, 7, 5],
        'C': [1, 2, 3, 1],
        'D': [9, 10, 11, 9]}

df = pd.DataFrame(data)

# 행 중복 여부 확인
duplicated_rows = df.duplicated()

# 중복된 행 제거
df_filtered = df[~duplicated_rows]

print(df_filtered)
```
