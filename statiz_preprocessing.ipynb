### 시즌중 트레이드된 선수 저장

```python
db = []
position = ['P', 'C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'DH']
for idx, row in df.iterrows():
    if row['팀'][-1] in position and len(row['팀']) - 3 > 1:
        db.append((row['시즌'], row['이름'],row['팀']))
    elif row['팀'][-2:] in position and len(row['팀']) - 4 > 1:
        db.append((row['시즌'], row['이름'],row['팀']))
```

```python
db = list(set(db))
db.sort(key = lambda x : (x[0], x[1]))
```

# 트레이드된 선수 정보 다시 받기

```python
from tqdm import tqdm
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
same_names = []

# 스탯티즈 사이트 접속
url = 'http://www.statiz.co.kr/stat.php'
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(10)
wait = WebDriverWait(driver, 10)
```

우선 동명이인은 따로 처리한다.

```python
trade_df = pd.DataFrame()
for year, name, info in tqdm(db):
    wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div/header/nav[1]/div/div[2]/form/div/input'))).send_keys(name)
    wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/header/nav[1]/div/div[2]/form/div/span/button'))).send_keys(Keys.ENTER)

    if len(pd.read_html(driver.page_source)[0]) > 1:
        same_names.append(name)
    else:
        wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/section[2]/div/div[1]/div/div[3]/div/div[2]/table/tbody/tr/td/a[2]'))).send_keys(Keys.ENTER)
        wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div[1]/div[1]/a[1]'))).send_keys(Keys.ENTER)

        tmp_df = pd.read_html(driver.page_source)[1]
        tmp_df.columns = tmp_df.columns.droplevel()
        trade_idx = tmp_df[tmp_df['연도'] == str(year)].index[0]
        tmp_df = tmp_df.iloc[[trade_idx+1, trade_idx+2]]
        tmp_df = tmp_df[tmp_df['연도'].isnull()]
        tmp_df.insert(0, '이름', name)
        tmp_df.insert(0, '시즌', year)
        trade_df = pd.concat([trade_df, tmp_df])
```

```python
def get_team(df):
    answer = ''
    if df['소속팀'] == '두산+':
        if df['시즌'] <= 1998:
            answer = 'OB 베어스'
        else:
            answer = '두산 베어스'
            
    elif df['소속팀'] == 'LG+':
        if df['시즌'] <= 1989:
            answer = 'MBC 청룡'
        else:
            answer = 'LG 트윈스'

    elif df['소속팀'] == "KIA+":
        if df['시즌'] <= 2000:
            answer = '해태 타이거즈'
        elif df['시즌'] == 2001:
            answer = '해태/KIA'
        else:
            answer = 'KIA 타이거즈'

    elif df['소속팀'] == '한화+':
        if df['시즌'] <= 1993:
            answer = '빙그레 이글스'
        else:
            answer = '한화 이글스'

    elif df['소속팀'] == '현대+':
        if df['시즌'] <= 1984:
            answer = '삼미 슈퍼스타즈'
        elif df['시즌'] == 1985:
            answer = '삼미/청보'
        elif df['시즌'] <= 1987:
            answer = '청보 핀토스'
        elif df['시즌'] <= 1995:
            answer = '태평양 돌핀스'
        else:
            answer = '현대 유니콘스'
        
    elif df['소속팀'] == 'SSG.SK':
        if df['시즌'] <= 2020:
            answer = 'SK 와이번스'
        else:
            answer = 'SSG 랜더스'

    elif df['소속팀'] == '히어로':
        if df['시즌'] == 2008:
            answer = '우리 히어로즈'
        elif df['시즌'] == 2009:
            answer = '히어로즈'
        elif df['시즌'] <= 2018:
            answer = '넥센 히어로즈'
        else:
            answer = '키움 히어로즈'
    else:
        answer = df['소속팀']

    return answer
```

```python
df['소속팀'] = df.apply(get_team, axis=1)
```

```python
def get_position(df):
    answer = ''
    if df['팀'][-1] in position:
        answer = df['팀'][-1]
        
    elif df['팀'][-2:] in position:
        answer = df['팀'][-2:]
    else:
        answer = np.NaN
    return answer
```

```python
df['포지션'] = df.apply(get_position, axis=1)
```

```python
def fix_team(team):
    answer = ''
    if team == '롯데':
        answer = '롯데 자이언츠'
    elif team == '삼성':
        answer = '삼성 라이온즈'
    elif team == 'NC':
        answer = 'NC 다이노스'
    elif team == 'kt' or team == 'KT':
        answer = 'KT 위즈'
    elif team == '쌍방울':
        answer = '쌍방울 레이더스'
    elif team == 'SK':
        answer = 'SK 와이번스'
    elif team == 'LG':
        answer = 'LG 트윈스'
    elif team == 'KIA':
        answer = 'KIA 타이거즈'
    elif team == '한화':
        answer = '한화 이글스'
    elif team == '두산':
        answer = '두산 베어스'
    elif team == '해태':
        answer = '해태 타이거즈'
    elif team == '현대':
        answer = '현대 유니콘스'
    elif team == '넥센':
        answer = '넥센 히어로즈'
    elif team == '키움':
        answer = '키움 히어로즈'
    elif team == 'SSG':
        answer = 'SSG 랜더스'
    elif team == 'OB':
        answer = 'OB 베어스'
    elif team == '태평양':
        answer = '태평양 돌핀스'
    elif team == 'MBC':
        answer = 'MBC 청룡'
    elif team == '빙그레':
        answer = '빙그레 이글스'
    else: 
        answer = team

    return answer
```

```python
output_df['소속팀'] = output_df['소속팀'].apply(fix_team)
```
